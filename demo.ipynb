```typescript
import matplotlib.pyplot as plt

data = [12, 19, 8, 15, 22, 18, 25]
plt.figure(figsize=(8, 4))
plt.plot(data, marker='o', color='#D97706', linewidth=2)
plt.title("User Activity Q1")
plt.ylabel("Weekly activity")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```typescript
# Equivalent to console.table
data = [
    {"method": "GET", "endpoint": "/api/users", "ms": 12},
    {"method": "POST", "endpoint": "/api/login", "ms": 45},
    {"method": "DELETE", "endpoint": "/api/session", "ms": 8},
]

header = f"{'method':<8} | {'endpoint':<15} | {'ms':>4}"
sep = "-" * len(header)
rows = "\n".join(f"{r['method']:<8} | {r['endpoint']:<15} | {r['ms']:>4}" for r in data)
print(header)
print(sep)
print(rows)
```

```typescript
import time

start = time.perf_counter()
total = 0
for i in range(10_000_000):
    total += i
end = time.perf_counter()

print(f"10M iterations: {(end - start) * 1000:.2f}ms")
print(f"Sum: {total}")
```
